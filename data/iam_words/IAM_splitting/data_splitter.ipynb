```python
import csv

train_A = "IAM-A/train.txt"
train_B = "IAM-B/train.txt"
train_C = "IAM-C/train.txt"
train_D = "IAM-D/train.txt"

valid_A = "IAM-A/valid.txt"
valid_B = "IAM-B/valid.txt"
valid_C = "IAM-C/valid.txt"
valid_D = "IAM-D/valid.txt"

test_A = "IAM-A/test.txt"
test_B = "IAM-B/test.txt"
test_C = "IAM-C/test.txt"
test_D = "IAM-D/test.txt"
```

```python
with open("train.txt", "w") as outfile:
    for fname in [train_A, train_B, train_C, train_D]:
        with open(fname) as infile:
            for line in infile:
                outfile.write(line)
with open("valid.txt", "w") as outfile:
    for fname in [valid_A, valid_B, valid_C, valid_D]:
        with open(fname) as infile:
            for line in infile:
                outfile.write(line)
with open("test.txt", "w") as outfile:
    for fname in [test_A, test_B, test_C, test_D]:
        with open(fname) as infile:
            for line in infile:
                outfile.write(line)
```

```python
import csv

ref_file_path = "../iam_words/label.csv"

with open(train_B, 'r') as in_file, open("IAM-B/train.csv", 'w') as out_file, open(ref_file_path) as ref_file:
    count = 0
    writer = csv.writer(out_file, delimiter='\t')
    writer.writerow(["No", "Name", "Label"])
    ref_reader = csv.reader(ref_file, delimiter='\t')
    
    for line in in_file:
        for row in ref_reader:
            if str(line.strip()) in row[1]:
                count += 1
                writer.writerow([count, row[1], row[2]])
        ref_file.seek(0)  # Reset the file pointer to the beginning for each line in train.txt
with open(train_C, 'r') as in_file, open("IAM-C/train.csv", 'w') as out_file, open(ref_file_path) as ref_file:
    count = 0
    writer = csv.writer(out_file, delimiter='\t')
    writer.writerow(["No", "Name", "Label"])
    ref_reader = csv.reader(ref_file, delimiter='\t')
    
    for line in in_file:
        for row in ref_reader:
            if str(line.strip()) in row[1]:
                count += 1
                writer.writerow([count, row[1], row[2]])
        ref_file.seek(0)  # Reset the file pointer to the beginning for each line in train.txt
with open(train_D, 'r') as in_file, open("IAM-D/train.csv", 'w') as out_file, open(ref_file_path) as ref_file:
    count = 0
    writer = csv.writer(out_file, delimiter='\t')
    writer.writerow(["No", "Name", "Label"])
    ref_reader = csv.reader(ref_file, delimiter='\t')
    
    for line in in_file:
        for row in ref_reader:
            if str(line.strip()) in row[1]:
                count += 1
                writer.writerow([count, row[1], row[2]])
        ref_file.seek(0)  # Reset the file pointer to the beginning for each line in train.txt


with open(valid_A, 'r') as in_file, open("IAM-A/valid.csv", 'w') as out_file, open(ref_file_path) as ref_file:
    count = 0
    writer = csv.writer(out_file, delimiter='\t')
    writer.writerow(["No", "Name", "Label"])
    ref_reader = csv.reader(ref_file, delimiter='\t')
    
    for line in in_file:
        for row in ref_reader:
            if str(line.strip()) in row[1]:
                count += 1
                writer.writerow([count, row[1], row[2]])
        ref_file.seek(0)  # Reset the file pointer to the beginning for each line in train.txt
with open(valid_B, 'r') as in_file, open("IAM-B/valid.csv", 'w') as out_file, open(ref_file_path) as ref_file:
    count = 0
    writer = csv.writer(out_file, delimiter='\t')
    writer.writerow(["No", "Name", "Label"])
    ref_reader = csv.reader(ref_file, delimiter='\t')
    
    for line in in_file:
        for row in ref_reader:
            if str(line.strip()) in row[1]:
                count += 1
                writer.writerow([count, row[1], row[2]])
        ref_file.seek(0)  # Reset the file pointer to the beginning for each line in train.txt
with open(valid_C, 'r') as in_file, open("IAM-C/valid.csv", 'w') as out_file, open(ref_file_path) as ref_file:
    count = 0
    writer = csv.writer(out_file, delimiter='\t')
    writer.writerow(["No", "Name", "Label"])
    ref_reader = csv.reader(ref_file, delimiter='\t')
    
    for line in in_file:
        for row in ref_reader:
            if str(line.strip()) in row[1]:
                count += 1
                writer.writerow([count, row[1], row[2]])
        ref_file.seek(0)  # Reset the file pointer to the beginning for each line in train.txt
with open(valid_D, 'r') as in_file, open("IAM-D/valid.csv", 'w') as out_file, open(ref_file_path) as ref_file:
    count = 0
    writer = csv.writer(out_file, delimiter='\t')
    writer.writerow(["No", "Name", "Label"])
    ref_reader = csv.reader(ref_file, delimiter='\t')
    
    for line in in_file:
        for row in ref_reader:
            if str(line.strip()) in row[1]:
                count += 1
                writer.writerow([count, row[1], row[2]])
        ref_file.seek(0)  # Reset the file pointer to the beginning for each line in train.txt


with open(test_A, 'r') as in_file, open("IAM-A/test.csv", 'w') as out_file, open(ref_file_path) as ref_file:
    count = 0
    writer = csv.writer(out_file, delimiter='\t')
    writer.writerow(["No", "Name", "Label"])
    ref_reader = csv.reader(ref_file, delimiter='\t')
    
    for line in in_file:
        for row in ref_reader:
            if str(line.strip()) in row[1]:
                count += 1
                writer.writerow([count, row[1], row[2]])
        ref_file.seek(0)  # Reset the file pointer to the beginning for each line in train.txt
with open(test_B, 'r') as in_file, open("IAM-B/test.csv", 'w') as out_file, open(ref_file_path) as ref_file:
    count = 0
    writer = csv.writer(out_file, delimiter='\t')
    writer.writerow(["No", "Name", "Label"])
    ref_reader = csv.reader(ref_file, delimiter='\t')
    
    for line in in_file:
        for row in ref_reader:
            if str(line.strip()) in row[1]:
                count += 1
                writer.writerow([count, row[1], row[2]])
        ref_file.seek(0)  # Reset the file pointer to the beginning for each line in train.txt
with open(test_C, 'r') as in_file, open("IAM-C/test.csv", 'w') as out_file, open(ref_file_path) as ref_file:
    count = 0
    writer = csv.writer(out_file, delimiter='\t')
    writer.writerow(["No", "Name", "Label"])
    ref_reader = csv.reader(ref_file, delimiter='\t')
    
    for line in in_file:
        for row in ref_reader:
            if str(line.strip()) in row[1]:
                count += 1
                writer.writerow([count, row[1], row[2]])
        ref_file.seek(0)  # Reset the file pointer to the beginning for each line in train.txt
with open(test_D, 'r') as in_file, open("IAM-D/test.csv", 'w') as out_file, open(ref_file_path) as ref_file:
    count = 0
    writer = csv.writer(out_file, delimiter='\t')
    writer.writerow(["No", "Name", "Label"])
    ref_reader = csv.reader(ref_file, delimiter='\t')
    
    for line in in_file:
        for row in ref_reader:
            if str(line.strip()) in row[1]:
                count += 1
                writer.writerow([count, row[1], row[2]])
        ref_file.seek(0)  # Reset the file pointer to the beginning for each line in train.txt
```
